```python
import pandas as pd
```

```python
featuresACvsVG=pd.read_excel('Comparison_ACvsVG_Zoo.xlsx',sheet_name='Relaciones entre indicadores')
```

```python
featuresACvsVG.columns
```

```python
sel_cols=['selected_game_features','skill_feature','imSkill','fk_skill_level','id_skill_feature','game_feature','main_resource','filter_criteria']



def select_columns_df(df, col_list):
  new_df=df[col_list]
  return(new_df)


filtered_game_features=select_columns_df(featuresACvsVG,sel_cols)
filtered_game_features.dropna()
```

```python
filtered_game_features['filter_criteria'].unique()   #(subset='filter_criteria')
```

```python
gf_criteria=filtered_game_features.groupby('filter_criteria').count()
gf_criteria
```

```python
gf_time=filtered_game_features[filtered_game_features['filter_criteria']=='T']
gf_time
```
